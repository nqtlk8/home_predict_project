```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/raw/train.csv")
```

```python
garage_col = [col for col in df.columns if 'Garage' in col]
df_garage = df[garage_col]
print(df_garage.info())
```

```python
df_garage.isnull().sum()
```

```python
# Nếu GarageCars và GarageArea = 0 thì fill null bằng 'No Garage'
df_garage['GarageType'] = df_garage.apply(lambda row: 'No Garage' if row['GarageCars'] == 0 and row['GarageArea'] == 0 else row['GarageType'], axis=1)
df_garage['GarageFinish'] = df_garage.apply(lambda row: 'No Garage' if row['GarageCars'] == 0 and row['GarageArea'] == 0 else row['GarageFinish'], axis=1)
df_garage['GarageQual'] = df_garage.apply(lambda row: 'No Garage' if row['GarageCars'] == 0 and row['GarageArea'] == 0 else row['GarageQual'], axis=1)
df_garage['GarageCond'] = df_garage.apply(lambda row: 'No Garage' if row['GarageCars'] == 0 and row['GarageArea'] == 0 else row['GarageCond'], axis=1)
df_garage['GarageYrBlt'] = df_garage.apply(lambda row: 0 if row['GarageCars'] == 0 and row['GarageArea'] == 0 else row['GarageYrBlt'], axis=1)
print(df_garage.isnull().sum())
```
