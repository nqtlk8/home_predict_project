```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/raw/train.csv")
```

```python
pool_col = [col for col in df.columns if 'Pool' in col]
df_pool = df[pool_col]
print(df_pool.info())
```

```python
df_pool.isnull().sum()
```

```python
# kiểm tra nếu pool area = 0 thì pool quality = 'No Pool'
df_pool['PoolQC'] = df_pool.apply(lambda row: 'No Pool' if row['PoolArea'] == 0 else row['PoolQC'], axis=1)
print(df_pool.isnull().sum())
```

```python
misc_col = [col for col in df.columns if 'Misc' in col]
df_misc = df[misc_col]
print(df_misc.info())
```

```python
# nếu miscval = 0 thì miscfeature = 'No Misc'
df_misc['MiscFeature'] = df_misc.apply(lambda row: 'No Misc' if row['MiscVal'] == 0 else row['MiscFeature'], axis=1)
print(df_misc.isnull().sum())
```
