```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/raw/train.csv")
```

```python
bsmt_col = [col for col in df.columns if 'Bsmt' in col]
bsmt_data = df[bsmt_col]
print(bsmt_data.info())
```

```python
bsmt_null = bsmt_data.isnull()
print(bsmt_null.sum())
```

```python
# --- Các cột có Qual null, area bằng 0 tức là không có tầng hầm ---
# Fill null basement quality with "No Basement"
bsmt_qual_null = bsmt_data['BsmtQual'].isnull()
# nếu cột bsmt là No Basement các cột khác cũng fillnull tương tự
bsmt_data.loc[bsmt_qual_null, 'BsmtQual'] = 'No Basement'
bsmt_data.loc[bsmt_qual_null, 'BsmtCond'] = 'No Basement'
bsmt_data.loc[bsmt_qual_null, 'BsmtExposure'] = 'No Basement'
bsmt_data.loc[bsmt_qual_null, 'BsmtFinType1'] = 'No Basement'
bsmt_data.loc[bsmt_qual_null, 'BsmtFinType2'] = 'No Basement'
bsmt_data.loc[bsmt_qual_null, 'BsmtFinSF1'] = 0
bsmt_data.loc[bsmt_qual_null, 'BsmtFinSF2'] = 0
bsmt_data.loc[bsmt_qual_null, 'BsmtUnfSF'] = 0
bsmt_data.loc[bsmt_qual_null, 'TotalBsmtSF'] = 0
bsmt_data.loc[bsmt_qual_null, 'BsmtFullBath'] = 0
bsmt_data.loc[bsmt_qual_null, 'BsmtHalfBath'] = 0
print(bsmt_data.info())
```

```python
bsmt_null = bsmt_data.isnull()
print(bsmt_null.sum())
```
